## **Acceso al modelo sin configuración**

```python
from ollama import chat

response = chat(model='tinyllama', messages=[
  {
    'role': 'user',
    'content': 'Why is the sky blue?',
  },
])
print(response.message.content)
```

```python
# Calculo de tokens usados en la respuesta
prompt_tokens = response.prompt_eval_count    #  tokens de entrada
completion_tokens = response.eval_count       # tokens de salida
total_tokens = prompt_tokens + completion_tokens

print(f"Prompt tokens: {prompt_tokens}")
print(f"Completion tokens: {completion_tokens}")
print(f"Total tokens: {total_tokens}")
```

## **Acceso al modelo con configuración**

### Parámetros relevantes:

- num_ctx: ventana de contexto usada en esa ejecución.
- temperature: creatividad/aleatoriedad (baja = más determinista).
- top_p: muestreo por probabilidad acumulada.
- repeat_penalty: reduce repeticiones.
- num_predict: máximo de tokens de salida.
- seed: reproducibilidad.

### Parámetros típicos en Ollama / LLMs


| Parámetro        | Qué controla                                   | Rango común          | Valor típico inicial |
|------------------|-----------------------------------------------|----------------------|----------------------|
| `temperature`    | Aleatoriedad / creatividad                    | 0.0 – 2.0            | 0.2 – 0.8            |
| `top_p`          | Muestreo por probabilidad acumulada           | 0.0 – 1.0            | 0.8 – 0.95           |
| `repeat_penalty` | Penaliza repeticiones                         | 1.0 – 2.0            | 1.05 – 1.2           |
| `num_ctx`        | Ventana de contexto utilizada                 | 512 – máximo modelo  | 2048 o 4096          |
| `num_predict`    | Máximo de tokens de salida                    | 1 – 4096+            | 128 – 512            |
| `seed`           | Reproducibilidad                              | Entero (0 – 2^31-1)  | 42 (o cualquier fijo)|

### Consideraciones

- `temperature = 0` → salida más determinista.  
- `num_ctx` no puede exceder el máximo que soporte el modelo.  
- Si se incrementa demasiado `repeat_penalty`, la respuesta puede volverse menos natural.


```python
opts = {
    "temperature": 0.2,
    "num_ctx": 4096,
    "top_p": 0.9,
    "repeat_penalty": 1.1,
    "seed": 42,
    "num_predict": 256
}

response = chat(
    model="tinyllama",
    messages=[{"role": "user", "content": "Why is the sky blue?"}],
    options=opts
)

print(response.message.content)
print("Opciones usadas:", opts)
```
