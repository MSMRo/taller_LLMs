```python
#import gradio as gr

def greet(name, intensity):
    return "Hello, " + name + "!" * int(intensity)

demo = gr.Interface(
    fn=greet,
    inputs=["text", "slider"],
    outputs=["text"],
    title="👋 Demo de Saludo",
    description="Escribe tu nombre y ajusta la intensidad del saludo.",
    api_name="predict"
)

demo.launch()
```

## Gradio y tinyllama

```python
import gradio as gr
from ollama import chat



opts = {
    "temperature": 0.2,
    "num_ctx": 4096,
    "top_p": 0.9,
    "repeat_penalty": 1.1,
    "seed": 42,
    "num_predict": 800
}

def fn_consulta_ollama(pregunta):
    global response
    response = chat(model='tinyllama',
    messages = [
        {
            'role': 'user',
            'content': pregunta,
        },
        ],
    options=opts)
    return response.message.content

demo = gr.Interface(
    fn=fn_consulta_ollama,
    inputs=["text"],
    outputs=["text"],
    title="Integración de Ollama con Gradio",
    description="Pregunta a Ollama.",
    api_name="predict"
)

demo.launch()
```

```python
prompt_tokens = response.prompt_eval_count    #  tokens de entrada
completion_tokens = response.eval_count       # tokens de salida
total_tokens = prompt_tokens + completion_tokens

print(f"Prompt tokens: {prompt_tokens}")
print(f"Completion tokens: {completion_tokens}")
print(f"Total tokens: {total_tokens}")
```
